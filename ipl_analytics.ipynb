```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')
matches.head()
deliveries.head()
```

```python
matches.head()
```

```python
matches.tail()
```

```python
# Check for missing values
print(matches.isnull().sum())
print(deliveries.isnull().sum())

# Drop rows with missing values in important columns
matches.dropna(subset=['winner'], inplace=True)
```

```python
#Total Matches Played Each Season
season_counts = matches['season'].value_counts().sort_index()
season_counts.plot(kind='bar', title='Total Matches Each Season')
plt.xlabel('Season')
plt.ylabel('Matches')
plt.show()
```

```python
#Most Successful Teams
matches['winner'].value_counts().head(10).plot(kind='bar', title='Most Successful Teams')
plt.xlabel('Team')
plt.ylabel('Matches Won')
plt.show()
```

```python
#Toss Winner vs Match Winner
toss_match_same = (matches['toss_winner'] == matches['winner']).value_counts()
toss_match_same.plot(kind='pie', autopct='%1.1f%%', labels=['Same', 'Different'], title='Toss Winner is Match Winner?')
plt.ylabel('')
plt.show()
```

```python
#Most Player of the Match Awards
matches['player_of_match'].value_counts().head(10).plot(kind='bar', title='Top 10 Players with Most POTM Awards')
plt.xlabel('Player')
plt.ylabel('Awards')
plt.show()
```

```python
#Top Run Scorers
top_batsmen = deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_batsmen.plot(kind='bar', title='Top 10 Run Scorers')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.show()
```

```python
#Most Fours and Sixes
fours = deliveries[deliveries['batsman_runs'] == 4]['batter'].value_counts().head(10)
sixes = deliveries[deliveries['batsman_runs'] == 6]['batter'].value_counts().head(10)

# Plot
fig, axs = plt.subplots(1, 2, figsize=(15,5))
fours.plot(kind='bar', ax=axs[0], title='Top 10 - Fours Hit')
sixes.plot(kind='bar', ax=axs[1], title='Top 10 - Sixes Hit')
plt.show()
```

```python
#Most Wickets Taken
# Filter out legal dismissals
wickets = deliveries[deliveries['dismissal_kind'].notna() & (deliveries['dismissal_kind'] != 'run out')]
top_bowlers = wickets['bowler'].value_counts().head(10)

top_bowlers.plot(kind='bar', title='Top 10 Wicket Takers')
plt.xlabel('Bowler')
plt.ylabel('Wickets')
plt.show()
```

```python
#Strike Rate of Top Batsmen
# Total runs and balls faced per batsman
batsman_stats = deliveries.groupby('batter').agg(
    runs_scored=('batsman_runs', 'sum'),
    balls_faced=('ball', 'count')
)

# Calculate strike rate
batsman_stats['strike_rate'] = (batsman_stats['runs_scored'] / batsman_stats['balls_faced']) * 100

# Filter out those who faced at least 200 balls
qualified_batsmen = batsman_stats[batsman_stats['balls_faced'] >= 200].sort_values(by='strike_rate', ascending=False)

# Display top 10
qualified_batsmen.head(10)
```

```python
#Bowler Economy Rate
# Filter out wides and no-balls using 'extras_type'
legal_deliveries = deliveries[
    (~deliveries['extras_type'].isin(['wides', 'noballs'])) | (deliveries['extras_type'].isna())
]


# Group by bowler from legal deliveries
bowler_stats = legal_deliveries.groupby('bowler').agg(
    total_runs=('total_runs', 'sum'),
    balls_bowled=('ball', 'count')
)

# Calculate overs and economy
bowler_stats['overs'] = bowler_stats['balls_bowled'] // 6 + (bowler_stats['balls_bowled'] % 6) / 6
bowler_stats['economy'] = bowler_stats['total_runs'] / bowler_stats['overs']

# Filter those with enough balls bowled
qualified_bowlers = bowler_stats[bowler_stats['balls_bowled'] >= 200].sort_values(by='economy')

# Show top 10 most economical bowlers
qualified_bowlers.head(10)

```

```python
 #Most Consistent Batsmen
# Runs per match
runs_per_match = deliveries.groupby(['match_id', 'batter'])['batsman_runs'].sum().reset_index()
player_consistency = runs_per_match.groupby('batter').agg(
    matches_played=('match_id', 'count'),
    total_runs=('batsman_runs', 'sum')
)
player_consistency['avg_runs_per_match'] = player_consistency['total_runs'] / player_consistency['matches_played']
player_consistency[player_consistency['matches_played'] >= 30].sort_values(by='avg_runs_per_match', ascending=False).head(10)
```

```python
#Head-to-Head Matchups – Team Win Records

#Count Wins for Each Matchup
#Drop matches with no result
completed_matches = matches[matches['winner'].notna()]

#  consistent matchup key (Team A vs Team B regardless of order)
completed_matches['matchup'] = completed_matches.apply(
    lambda x: tuple(sorted([x['team1'], x['team2']])), axis=1
)

# Group by matchup and winner
head_to_head = completed_matches.groupby(['matchup', 'winner']).size().reset_index(name='wins')

# Expand the matchup tuple into two columns for readability
head_to_head[['Team A', 'Team B']] = pd.DataFrame(head_to_head['matchup'].tolist(), index=head_to_head.index)
head_to_head = head_to_head[['Team A', 'Team B', 'winner', 'wins']]
head_to_head.sort_values(by=['Team A', 'Team B', 'wins'], ascending=[True, True, False], inplace=True)

# Show top matchups
head_to_head.head(10)
```

```python
#in pivot table view
pivot = head_to_head.pivot_table(
    index=['Team A', 'Team B'],
    columns='winner',
    values='wins',
    fill_value=0
)

pivot.head(10)
```
